# 1) Crear series a partir de listas y a partir de arrays.

```python
import pandas as pd
import numpy as np

# Serie desde una lista

lista = [10, 20, 30, 40]
serieLista = pd.Series(lista)
print("Serie desde lista:")
print(serieLista)

# Serie desde un array de numpy

array = np.array([5, 15, 25, 35])
serieArray = pd.Series(array)
print("\nSerie desde array:")
print(serieArray)
```

# 2) Crear dataframes a partir de listas anidadas, y de arrays anidados.

```python
# DataFrame desde listas anidadas

ListasAnidadas = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
dfListas = pd.DataFrame(ListasAnidadas)
print("\nDataFrame desde listas anidadas:")
print(dfListas)
```

```python
# DataFrame desde arrays anidados

ArraysAnidados = np.array([[10, 20, 30], [40, 50, 60], [70, 80, 90]])
dfArrays = pd.DataFrame(ArraysAnidados)
print("\nDataFrame desde arrays anidados:")
print(dfArrays)
```

# 3) Crear dataframes especificando etiquetas para las columnas.


```python
# DataFrame con etiquetas para las columnas (utilizo nombres de meses para el ejemplo)

dfEtiquetas = pd.DataFrame(ArraysAnidados, columns=["Enero", "Febrero", "Marzo"])
print("\nDataFrame con etiquetas para las columnas:")
print(dfEtiquetas)
```

# 8) Seleccionar los datos de todas las mujeres de primera clase.

```python
# Filtro las filas donde 'Sex' es 'female' y 'Pclass' es 1 (primera clase)

MujeresPrimeraClase = Titanic[(Titanic['Sex'] == 'female') & (Titanic['Pclass'] == 1)]

MujeresPrimeraClase
```

# 9) Seleccionar los datos de los varones menores de edad de segunda clase.

```python
# En este caso, tomo como menores de edad a aquellos qe posean menos de 13 a√±os.

# Filtro las filas donde 'Sex' es 'male', 'Age' es menor a 13 y 'Pclass' es 2 (segunda clase)

Varones2daClase = Titanic[(Titanic['Sex'] == 'male') & (Titanic['Age'] < 13) & (Titanic['Pclass'] == 2)]

Varones2daClase
```

# 10) Encontrar el promedio de edad de hombres y de las mujeres

```python
# Calculo el promedio de edad para hombres y mujeres. Al usar "Group By" agrupo los g√©neros

PromedioEdad = Titanic.groupby('Sex')['Age'].mean()

PromedioEdad
```

# 11) Encontrar la cantidad de sobrevivientes de hombres y las mujeres.

```python
# Agrupo por 'Sex' y cuento la cantidad de sobrevivientes. En este df = Survived == 1

CantidadSobrevivientes = Titanic[Titanic['Survived'] == 1].groupby('Sex').size()

CantidadSobrevivientes
```
