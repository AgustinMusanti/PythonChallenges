# 1) Construir 3 gráficos de línea a partir de un array creado con la función linspace de numpy.

## a) Uno que represente una curva lineal.
## b) Otro que represente una curva cuadrática.
## c) Un tercero que represente una curva logarítmica.

```python
import numpy as np
import matplotlib.pyplot as plt

# Creo un array con linspace
x = np.linspace(0.1, 20, 200)

# Calculo los valores para cada curva

y_linear = x                  # Curva lineal
y_quadratic = x**2/50         # Curva cuadrática
y_logarithmic = np.log(x)     # Curva logarítmica

# Gráficos
plt.figure(figsize=(12, 8))

# Gráfico lineal (inciso a)

plt.subplot(3, 1, 1)
plt.plot(x, y_linear, label="Lineal", color="blue")
plt.title("Curva Lineal")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()

# Gráfico cuadrático (inciso b)

plt.subplot(3, 1, 2)
plt.plot(x, y_quadratic, label="Cuadrática", color="green")
plt.title("Curva Cuadrática")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()

# Gráfico logarítmico (inciso c)

plt.subplot(3, 1, 3)
plt.plot(x, y_logarithmic, label="Logarítmica", color="red")
plt.title("Curva Logarítmica")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()

# Ajusto los espacios usando tight_layout para evitar solapamientos

plt.tight_layout()
plt.show()
```

# 2) Para los gráficos realizados en el ejercicio anterior, hay que cambiar el formato. Además, los 3 gráficos deben de quedar en una sola figura juntos, hay que activar la grilla y ponerle el título a la figura “Gráfico de pruebas” y cambiar el tamaño del mismo para poder apreciarlo mejor, según corresponda. Agregar leyendas.

## a) El que representa la curva lineal, debe ser de color rojo y punteada ( · · · ), con marcador del tipo “square maker”.
## b) La curva exponencial, sólida de color negro y con un marcador del tipo “circle maker”.
## c) La curva logarítmica debe ser discontinua ( - - - ) de color amarillo y con marcador del tipo “x maker”.

```python
x = np.linspace(0.1, 20, 200)

y_linear = x                    # Curva lineal
y_exponential = np.exp(x / 10)  # Curva exponencial
y_logarithmic = np.log(x)       # Curva logarítmica

# Creo la figura

plt.figure(figsize=(14, 8))
plt.title("Gráfico de pruebas", fontsize=16)  # Título general

# Curva lineal (inciso a)

plt.plot(
    x, y_linear, 'r--s', label="Lineal",  # Rojo, punteada, marcador cuadrado
    markersize=6, linewidth=1
)

# Curva exponencial (inciso b)

plt.plot(
    x, y_exponential, 'k-o', label="Exponencial",  # Negro, sólida, marcador circular
    markersize=6, linewidth=1
)

# Curva logarítmica (inciso c)

plt.plot(
    x, y_logarithmic, 'y--x', label="Logarítmica",  # Amarillo, discontinua, marcador 'x'
    markersize=6, linewidth=1
)

plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.grid(True)  # Con esta linea activo la grilla
plt.legend(fontsize=12)  # Con esta linea muestro las leyendas

# Figura final

plt.tight_layout()
plt.show()
```

# 3) Se tiene una lista de listas de la evolución del dólar por año y un diccionario con la demanda según en porcentaje de los que compran dólares o hacen uso de la divisa en Argentina:
## Hacer un subplot con un gráfico de barras de dolarValor y una torta de dolarComp con sus respectivos títulos.
dolarValor = [[2020,70],[2021,95],[2022,110]]
(Año – Valor dólar oficial)
dolarComp = {“privados”:40,”particulares”:20,”estatales”:30,”otros”:10}

```python
# Datos del inciso

dolarValor = [[2020, 70], [2021, 95], [2022, 110]]  # Año y valor del dólar oficial
dolarComp = {"Privados": 40, "Particulares": 20, "Estatales": 30, "Otros": 10}  # Demanda en porcentaje

# Preparo los datos para el gráfico de barras

anios = [item[0] for item in dolarValor]  # Años
valores = [item[1] for item in dolarValor]  # Valores del dólar

# Preparo los datos para el gráfico de torta

categorias = list(dolarComp.keys())  # Categorías
porcentajes = list(dolarComp.values())  # Porcentajes

# Creo la figura

plt.figure(figsize=(14, 6))

# Subplot 1: Gráfico de barras

plt.subplot(1, 2, 1)  # Primer gráfico
plt.bar(anios, valores, color="skyblue", edgecolor="black")
plt.title("Evolución del dólar oficial", fontsize=14)
plt.xlabel("Año", fontsize=12)
plt.ylabel("Valor ($)", fontsize=12)
plt.xticks(anios, fontsize=10)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Subplot 2: Gráfico de torta

plt.subplot(1, 2, 2)  # Segundo gráfico
plt.pie(
    porcentajes,
    labels=categorias,
    autopct="%1.1f%%",
    startangle=90,
    colors=["gold", "lightgreen", "lightcoral", "lightskyblue"],
    explode=(0.1, 0, 0, 0)  # En esta linea busco resaltar 'privados'
)
plt.title("Demanda de dólares en Argentina", fontsize=14)

# Utilizo tight_layout para evitar solapamientos

plt.tight_layout()
plt.show()
```
